```python
print()

print()
print('Решение через функцию')
print()


def vozvrat_debt_family_st(row):
    ##пишем функцию, которая возвращает значение согласно следующему правилу:
    ###Заявитель с сем. положением "женат / замужем" , не имеющий невозвратов по кредиту - 
    ###статус'женат / замужем с хорошей кредитной историей'
    ###Заявитель с сем. положением "женат / замужем" , имеющий невозвраты по кредиту  - статус 
    ###'женат / замужем с плохой кредитной историей
    ###Заявитель с сем. положением "гражданский брак" , не имеющий невозвратов по кредиту - 
    ###статус 'гражданский брак с хорошей кредитной историей'
    ###Заявитель с сем. положением "гражданский брак" , имеющий невозвраты по кредиту  - статус 
    ###'гражданский брак с плохой кредитной историей'
    
       
     
    debt = row['debt']
    family_status = row['family_status_id']

    if debt == 0: #  женат / замужем нет невозвратов
        if family_status == 0: # женат / замужем
            return 'женат / замужем с хорошей кредитной историей'
        
    if debt == 1: #  есть невозврат кредита
        if family_status == 0: # женат / замужем
            return 'женат / замужем с плохой кредитной историей'    
        

    if debt == 0: #нет невозврата кредита
        if family_status == 1: #гражданский брак
            return 'гражданский брак с хорошей кредитной историей'
        
    if debt == 1: #есть невозврат кредита
        if family_status == 1:
            return 'гражданский брак с плохой кредитной историей'
          
      
    if debt == 0: #нет невозврата кредита
        if family_status == 2: #вдовец / вдова 
            return 'вдовец / вдова с хорошей кредитной историей'
        
    if debt == 1: #есть невозврат кредита
        if family_status == 2:
            return 'вдовец / вдова с плохой кредитной историей'
        
    if debt == 0: #нет невозврата кредита
        if family_status == 3: #в разводе
            return 'в разводе с хорошей кредитной историей'
        
    if debt == 1: #есть невозврат кредита
        if family_status == 3:
            return 'в разводе с плохой кредитной историей'
        
    if debt == 0: #нет невозврата кредита
        if family_status == 4: #не женат / не замужем
            return 'не женат / не замужем с хорошей кредитной историей'
    return 'не женат / не замужем с плохой кредитной историей'    

row_values = [0, 0] #значения наличия "кредитной истории" и "семейного положения"
row_columns = ['debt', 'family_status_id'] #названия столбцов

row = pd.Series(data=row_values, index=row_columns)
#print(vozvrat_debt_family_st(row)) # проверка работы функции
    
data['debt_family_st'] = data.apply(vozvrat_debt_family_st, axis=1)  
dep_debt_family_st = data['debt_family_st'].value_counts()

print(dep_debt_family_st)


# Создаем сводную таблицу через метод groupby                                                  
data_grouped_FS = data.groupby('family_status')[['debt']].agg(['count',"sum", "mean"])
#data_grouped_FS['ratio'] = data_grouped_FS["debt"]["sum"]/data_grouped_FS["debt"]["count"]

data_grouped_FS
```
